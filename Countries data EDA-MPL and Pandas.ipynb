#### Q2. What is the distribution of life expectancy across all countries and years?


```python
plt.figuresize=100
plt.hist(countries['lifeExpectancy'],
        facecolor='skyblue',
        edgecolor='black',
        bins=20)
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.title('Distribution of Life Expectancy')
plt.show()
```

#### Q3. How does life expectancy vary over the years for a specific country (e.g., Afghanistan)?
https://www.geeksforgeeks.org/matplotlib-markers-module-in-python/
https://www.geeksforgeeks.org/grids-in-matplotlib/


```python
plt.figure(figsize=(10,6))
afganistan=countries[countries['country']=='Afghanistan']
plt.plot(afganistan['year'],afganistan['lifeExpectancy'],
         color='green',
         marker='o')
plt.title('Life Expectancy Over The Years in Afghanistan')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
         
```

#### Q4. What is the relationship between GDP per capita and life expectancy?


```python
plt.scatter(countries['gdpPerCapita'],
            countries['lifeExpectancy'],
            color='red')
plt.title('GDP per Capita vs. Life Expectancy')
plt.xlabel("GDP per capita")
plt.ylabel("life expectancy")
plt.show()
```

#### Q5. How has the population changed over time for a specific continent (e.g., Africa)?

** Use groupby for year with populations

```python
africa=countries[countries['continent']=='Africa']
group_data=africa.groupby('year')['population'].sum()
plt.plot(group_data,
         color='red',
         marker='o')
plt.title('Population Over Time in Africa')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

#### Q6. What is the trend of GDP per capita over the years for a specific continent (e.g., Europe)?


```python
europe=countries[countries['continent']=='Europe']
group_data1=europe.groupby('year')['gdpPerCapita'].sum()
plt.plot(group_data1,
         color='blue',
         marker='o')
plt.title('GDP per Capita Trend Over Time in Europe')
plt.xlabel('Year')
plt.ylabel('GDP per Capita')
plt.grid(True)
plt.show()
```
